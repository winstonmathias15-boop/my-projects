```python
marks=input("Enter your mark out of 50 here: ")
grades=int(marks)
score=((grades/50)*100)
active= True

while active:
  if score <= 50:
    print (f"You got {score}%, you failed homey!")
  elif score > 100:
    print("That's not possible my guy!")
  else:
    print (f"You got {score}%, you passed damnt!")
  active=False
```
