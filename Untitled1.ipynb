```python
from enum import Enum, auto

class State(Enum):
  IDLE = auto()
  DOWNLOADING =auto()
  PAUSED = auto()
  COMPLETED = auto()

class Downloader():
  def __init__(self, url, current_state=State.IDLE):
    self.url = url
    self.current_state = State.IDLE

  def start_download(self):
    if self.current_state == State.IDLE:
      self.current_state = State.DOWNLOADING
      print("Downloading is starting...")
    else:
      print(f"Download cannot start, current state is {self.current_state.name}")

  def pause_download(self):
    if self.current_state == State.DOWNLOADING:
      self.current_state = State.PAUSED
      print("Download is paused...")
    else:
      print(f"Download cannot be paused, current state is {self.current_state.name}")

  def finish_download(self):
    if self.current_state == State.DOWNLOADING:
      self.current_state = State.COMPLETED
      print("Download is completed...")
    else:
      print(f"Download cannot be completed, current state is {self.current_state.name}")

manager = Downloader("https://www.google.com")
manager.start_download()
manager.pause_download()
manager.finish_download()
```
